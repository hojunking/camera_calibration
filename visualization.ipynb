```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

# Load the .mat file
with h5py.File(r"C:\Users\hojun\git\data\nyu_depth_v2_labeled.mat", 'r') as file:
    depths = np.array(file['depths'])
    rgbs = np.array(file['images'])

# Just as an example, take the first depth and RGB image from the dataset
depth_image = np.transpose(depths[:,:,0], (1, 0))
rgb_image = np.transpose(rgbs[:,:,:,0], (1, 2, 0))
```

```python
rows, cols = depth_image.shape
c, r = np.meshgrid(np.arange(cols), np.arange(rows), sparse=True)
z = depth_image
x = (c - 319.5) * z / 525.0
y = (r - 239.5) * z / 525.0
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(x, cmap='jet')
ax[0].set_title('X values')
ax[1].imshow(y, cmap='jet')
ax[1].set_title('Y values')
ax[2].imshow(z, cmap='jet')
ax[2].set_title('Z (depth) values')
plt.show()
```
